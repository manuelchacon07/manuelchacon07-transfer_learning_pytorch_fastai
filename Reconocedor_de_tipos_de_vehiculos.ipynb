## Reconocedor de tipos de vehiculos (coche, moto o bus) usando transfer learning

Para crear este modelo he usado un modelo preentrenado llamado ```resnet18``` mas un conjunto de imagenes

[Acceso al repositorio](https://github.com/manuelchacon07/transfer_learning_pytorch_fastai)

````python
# Importamos las librerias necesarias

from fastai.vision.widgets import *
from fastbook import *
````

````python
# Introducimos la ruta y comprobamos si esta el modelo
path = Path('')
path.ls(file_exts='.pkl')
````

````python
# Cargamos el modelo
model = load_learner(path/'TL_tipos_vehiculos.pkl')
````

````python
# Variable para generar un boton que guarda la subida
btn_upload = widgets.FileUpload(description='Subir')

# Variable para sacar la imagen subida
out_pl = widgets.Output()

# Variable para generar un boton que ejecuta la funcion "on_click_classify"
btn_run = widgets.Button(description='Clasificar')

# Variable que almacena la etiqueta de la prediccion
lbl_pred = widgets.Label()

# Funcion clasificador
def on_click_classify(change):
    # Mostramos por pantalla la foto elegida
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(512,512))
    
    # Realizamos la prediccion y lo guardamos en variables
    pred,pred_idx,probs = model.predict(img)

    # Modificamos la etiqueta de prediccion
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

# Damos vida al boton "btn_run"
btn_run.on_click(on_click_classify)
````

````python
# Crea el panel de nuestra demo
VBox([widgets.Label('Sube tu vehiculo a comprobar'), 
      btn_upload, btn_run, out_pl, lbl_pred])
````
